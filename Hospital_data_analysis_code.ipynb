```python
import re

def standardize_blood_type(bt):
    if pd.isna(bt):
        return bt
    bt = str(bt).strip().upper()
    # Remove any character that is not A, B, O, +, -
    bt = re.sub(r'[^ABO+\-]+', '', bt)

    if not bt: # Handle empty strings after cleaning
        return None

    # Handle common incorrect representations
    replacements = {
        'B-+': 'B+', 'AB+-': 'AB+', 'A+-': 'A+', 'AB-+': 'AB+', 'B+-': 'B+',
        'B++': 'B+', 'A--': 'A-', 'AB++': 'AB+', 'O++': 'O+', 'A-+': 'A+',
        'B--': 'B-', 'AB--': 'AB-', 'O--': 'O-', 'O+-': 'O+', 'O-+': 'O-',
        'A +': 'A+', 'A -': 'A-', 'B +': 'B+', 'B -': 'B-',
        'AB +': 'AB+', 'AB -': 'AB-', 'O +': 'O+', 'O -': 'O-',
    }
    for old, new in replacements.items():
        bt = bt.replace(old, new)

    # Further simplification: if multiple + or -
    bt = re.sub(r'\++', '+', bt)
    bt = re.sub(r'-+', '-', bt)

    # Final check for valid blood types (A, B, AB, O with +/-)
    if bt in ['A+', 'A-', 'B+', 'B-', 'AB+', 'AB-', 'O+', 'O-']:
        return bt
    else:
        # For anything still not standard, set to 'UNKNOWN_BLOOD_TYPE'
        return 'UNKNOWN_BLOOD_TYPE'

df['Blood Type'] = df['Blood Type'].apply(standardize_blood_type)

df['Insurance Provider'] = df['Insurance Provider'].str.replace("H"," H")
df
```
